# Imports

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

# Importing Data

```python
file = open("Data/ClusterData.csv", "r")
data = []
for line in file:
    # skip the header line
    if 'bin' in line:
        continue
    # turns each row into a list then appends to data
    data.append(list(map(float, line.rstrip().split(','))))
```

# 2D Plotting

```python
# creates a dictionary for easy access
sets = {}

# fill the set
sets['X'] = [row[0] for row in data]
sets['Y'] = [row[1] for row in data]
sets['Z'] = [row[2] for row in data]
# binary column
sets['B'] = [row[3] for row in data]
    
# iterates on the string keys associated with each set
for i in sets:
    for q in sets:
        
        # so we do not plot a set against itself or binary
        if i == q or i == 'B' or q == 'B':
            continue
         
        # creates each plot. Dictionary key allows for easy string manipulation
        # play around with the coefficient on variable s (1/100 makes clusters viewable)
        plt.scatter(sets[i], sets[q], s = np.pi/100, c= sets['B'])
        plt.xlabel(i)
        plt.ylabel(q)
        plt.title("Plot of " + q + " Against " + i)
        plt.show()
```

# 3D Plotting

```python
# create the 3D figure using the sets above
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# add attributes
ax.scatter(sets['X'], sets['Y'], sets['Z'], c = sets['B'], s = np.pi/100)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
